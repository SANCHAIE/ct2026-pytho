**วัตถุประสงค์การเรียนรู้ (Learning Objectives)**

เมื่อศึกษาบทนี้จบแล้ว ผู้เรียนสามารถ:
1. ตรวจสอบและเปรียบเทียบโค้ดของตนเองกับเฉลยที่แนะนำได้
2. ทบทวนและสรุปทักษะที่เรียนรู้ตลอดทั้งหลักสูตรได้
3. วางแผนเส้นทางการเรียนรู้ต่อในขั้นถัดไปได้

> **คำแนะนำการใช้งาน:** ลองทำแบบฝึกหัดในบทที่ 3 ด้วยตนเองก่อน แล้วค่อยเปิดดูเฉลยในบทนี้


```python
# === เตรียมข้อมูล (เหมือนคลิป 3) ===
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (10, 5)

np.random.seed(2026)
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
seasonal = [1.8, 2.0, 1.6, 1.0, 0.6, 0.4, 0.3, 0.4, 0.5, 0.7, 1.0, 1.5]
provinces_info = [
    ("Bangkok",42,"Central"),("ChiangMai",65,"North"),("ChiangRai",58,"North"),
    ("KhonKaen",35,"Northeast"),("UdonThani",32,"Northeast"),("Chonburi",25,"East"),
    ("Phuket",15,"South"),("Songkhla",18,"South"),("NakhonRatchasima",38,"Northeast"),
    ("SamutPrakan",48,"Central")
]
rows = []
for mi, month in enumerate(months):
    for prov, base_pm, region in provinces_info:
        pm25 = max(5, int(base_pm * seasonal[mi] + np.random.randint(-8, 8)))
        temp = round(25 + seasonal[mi]*5 + np.random.uniform(-2, 3), 1)
        humidity = int(60 + (1 - seasonal[mi]) * 25 + np.random.randint(-5, 5))
        aqi = int(pm25 * 2.1 + np.random.randint(-5, 5))
        rows.append({"month":month, "province":prov, "region":region, "pm25":pm25, "temp_c":temp, "humidity":humidity, "aqi":aqi})
for idx in [7, 23, 45, 67, 89, 102]:
    if idx < len(rows): rows[idx]["pm25"] = np.nan
for idx in [15, 38, 71]:
    if idx < len(rows): rows[idx]["temp_c"] = np.nan
df = pd.DataFrame(rows)

# Clean
df_clean = df.copy()
df_clean["pm25"] = df_clean["pm25"].fillna(df_clean["pm25"].mean())
df_clean["temp_c"] = df_clean["temp_c"].fillna(df_clean["temp_c"].mean())
print("Data ready!")
```

---
## เฉลย Level 1 — เติมโค้ดในช่องว่าง

เนื้อหาด้านล่างแสดงคำตอบที่ถูกต้องสำหรับทุกช่องว่าง `___` พร้อมคำอธิบาย


```python
# L1 ข้อ 1.2: เติมค่าที่หาย
df_clean["pm25"] = df_clean["pm25"].fillna(df_clean["pm25"].mean())
df_clean["temp_c"] = df_clean["temp_c"].fillna(df_clean["temp_c"].mean())

# L1 ข้อ 2.1: groupby("province")
pm_by_province = df_clean.groupby("province")["pm25"].mean().sort_values(ascending=False)
print(pm_by_province)

# L1 ข้อ 3.1: .values
# plt.bar(..., pm_by_province.values, ...)
# plt.ylabel('Average PM2.5')

# L1 ข้อ 4.1: prov (ตัวแปรที่ loop)
# prov_data = df_clean[df_clean["province"] == prov]

# L1 ข้อ 5.1: * 100
# unsafe_pct = len(unsafe_months) / len(df_clean) * 100
```

## เฉลย Level 2 — โจทย์พร้อมคำใบ้

เนื้อหาด้านล่างแสดงตัวอย่างโค้ดแนะนำสำหรับแต่ละโจทย์ ผู้เรียนอาจมีแนวทางการเขียนโค้ดที่แตกต่างออกไปได้ตราบที่ผลลัพธ์ถูกต้อง


```python
# L2 ข้อ 1: PM2.5 เฉลี่ยตามภาค + Bar Chart แนวนอน
pm_by_region = df_clean.groupby("region")["pm25"].mean().sort_values()

plt.figure(figsize=(8, 4))
colors = ['red' if pm > 50 else 'orange' if pm > 25 else 'green'
          for pm in pm_by_region.values]
plt.barh(pm_by_region.index, pm_by_region.values, color=colors)
plt.axvline(x=50, color='red', linestyle='--', alpha=0.5)
plt.xlabel('Average PM2.5')
plt.title('Average PM2.5 by Region')
plt.tight_layout()
plt.show()
```

```python
# L2 ข้อ 2: PM2.5 เฉลี่ยรายเดือน
months_order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
pm_by_month = df_clean.groupby("month")["pm25"].mean().reindex(months_order)

plt.figure(figsize=(10, 5))
colors = ['red' if pm > 50 else 'orange' if pm > 25 else 'green'
          for pm in pm_by_month.values]
plt.bar(pm_by_month.index, pm_by_month.values, color=colors)
plt.axhline(y=50, color='red', linestyle='--', alpha=0.5)
plt.ylabel('Average PM2.5')
plt.title('Monthly Average PM2.5 (All Provinces)')
plt.tight_layout()
plt.show()

print(f"เดือนที่แย่สุด: {pm_by_month.idxmax()} (PM2.5 = {pm_by_month.max():.1f})")
print(f"เดือนที่ดีสุด: {pm_by_month.idxmin()} (PM2.5 = {pm_by_month.min():.1f})")
```

```python
# L2 ข้อ 3: Scatter Plot แยกสีตามภาค
plt.figure(figsize=(10, 6))
region_colors = {"North": "red", "Central": "orange", "Northeast": "blue",
                 "East": "green", "South": "purple"}

for region in df_clean["region"].unique():
    data = df_clean[df_clean["region"] == region]
    plt.scatter(data["temp_c"], data["pm25"], label=region,
                color=region_colors.get(region, "gray"), alpha=0.5, s=30)

plt.xlabel('Temperature (C)')
plt.ylabel('PM2.5')
plt.title('Temperature vs PM2.5 by Region')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# L2 ข้อ 4: ครึ่งปีแรก vs ครึ่งปีหลัง
first_half = ["Jan","Feb","Mar","Apr","May","Jun"]
second_half = ["Jul","Aug","Sep","Oct","Nov","Dec"]

h1 = df_clean[df_clean["month"].isin(first_half)].groupby("province")["pm25"].mean()
h2 = df_clean[df_clean["month"].isin(second_half)].groupby("province")["pm25"].mean()

comparison = pd.DataFrame({"H1 (Jan-Jun)": h1, "H2 (Jul-Dec)": h2})
comparison["diff"] = comparison["H1 (Jan-Jun)"] - comparison["H2 (Jul-Dec)"]
comparison = comparison.sort_values("diff", ascending=False)

print("PM2.5 เปรียบเทียบครึ่งปีแรก vs ครึ่งปีหลัง")
print(comparison.round(1))
print("\n(diff > 0 = ครึ่งปีแรกแย่กว่า)")

# Bar Chart เปรียบเทียบ
x = range(len(comparison))
plt.figure(figsize=(12, 5))
plt.bar([i - 0.2 for i in x], comparison["H1 (Jan-Jun)"], width=0.4, label='H1 (Jan-Jun)', color='#e74c3c')
plt.bar([i + 0.2 for i in x], comparison["H2 (Jul-Dec)"], width=0.4, label='H2 (Jul-Dec)', color='#2ecc71')
plt.xticks(list(x), comparison.index, rotation=45)
plt.ylabel('Average PM2.5')
plt.title('PM2.5: First Half vs Second Half of Year')
plt.legend()
plt.tight_layout()
plt.show()
```
