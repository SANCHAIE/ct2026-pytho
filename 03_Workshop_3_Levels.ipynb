## ส่วนที่ 1 — เตรียมข้อมูล (ทุกระดับใช้ร่วมกัน)

**คำแนะนำ:** รันเซลล์ส่วนนี้ก่อนเสมอ เพื่อโหลดชุดข้อมูลที่ใช้ในการวิเคราะห์

ชุดข้อมูลนี้เป็นข้อมูลคุณภาพอากาศ PM2.5 รายเดือน จำนวน **10 จังหวัด** ตลอดทั้งปี รวม 120 แถว พร้อมค่าที่หายไป (missing values) จำลองสถานการณ์ข้อมูลจริง


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (10, 5)

# สร้าง Dataset: PM2.5 รายเดือน 10 จังหวัด (จำลองข้อมูลสมจริง)
np.random.seed(2026)

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
seasonal = [1.8, 2.0, 1.6, 1.0, 0.6, 0.4, 0.3, 0.4, 0.5, 0.7, 1.0, 1.5]

provinces_info = [
    ("Bangkok", 42, "Central"),
    ("ChiangMai", 65, "North"),
    ("ChiangRai", 58, "North"),
    ("KhonKaen", 35, "Northeast"),
    ("UdonThani", 32, "Northeast"),
    ("Chonburi", 25, "East"),
    ("Phuket", 15, "South"),
    ("Songkhla", 18, "South"),
    ("NakhonRatchasima", 38, "Northeast"),
    ("SamutPrakan", 48, "Central"),
]

rows = []
for mi, month in enumerate(months):
    for prov, base_pm, region in provinces_info:
        pm25 = max(5, int(base_pm * seasonal[mi] + np.random.randint(-8, 8)))
        temp = round(25 + seasonal[mi]*5 + np.random.uniform(-2, 3), 1)
        humidity = int(60 + (1 - seasonal[mi]) * 25 + np.random.randint(-5, 5))
        aqi = int(pm25 * 2.1 + np.random.randint(-5, 5))
        rows.append({
            "month": month, "province": prov, "region": region,
            "pm25": pm25, "temp_c": temp, "humidity": humidity, "aqi": aqi
        })

# เพิ่ม missing values จำลองข้อมูลจริง
for idx in [7, 23, 45, 67, 89, 102]:
    if idx < len(rows):
        rows[idx]["pm25"] = np.nan
for idx in [15, 38, 71]:
    if idx < len(rows):
        rows[idx]["temp_c"] = np.nan

df = pd.DataFrame(rows)

print(f"✅ Dataset พร้อมใช้งาน!")
print(f"   ขนาด: {df.shape[0]} แถว x {df.shape[1]} คอลัมน์")
print(f"   คอลัมน์: {list(df.columns)}")
print(f"   ข้อมูลที่หายไป: pm25 = {df['pm25'].isna().sum()}, temp_c = {df['temp_c'].isna().sum()}")
df.head(10)
```
