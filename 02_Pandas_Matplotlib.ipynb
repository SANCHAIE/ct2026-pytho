## หัวข้อที่ 1 — pandas: โหลดและสำรวจข้อมูล
**เวลาโดยประมาณ: 12 นาที**

`pandas` คือไลบรารีหลักสำหรับการจัดการข้อมูลตาราง (tabular data) ใน Python โดยใช้โครงสร้างข้อมูลที่เรียกว่า **DataFrame** ซึ่งคล้ายกับ Excel spreadsheet


```python
import pandas as pd

# สร้าง DataFrame (ตาราง)
data = {
    "province": ["Bangkok", "ChiangMai", "Phuket", "KhonKaen", "Chonburi",
                 "Songkhla", "ChiangRai", "Nakhon Ratchasima", "UdonThani", "SamutPrakan"],
    "pm25": [55, 120, 25, 45, 30, 28, 95, 50, 42, 60],
    "temp": [35.2, 28.1, 31.5, 33.8, 30.0, 29.5, 26.3, 34.2, 32.1, 34.8],
    "humidity": [70, 55, 80, 60, 75, 82, 50, 58, 65, 72],
    "region": ["Central", "North", "South", "Northeast", "East",
               "South", "North", "Northeast", "Northeast", "Central"]
}

df = pd.DataFrame(data)
df
```

```python
# สำรวจข้อมูล
print(f"ขนาดข้อมูล: {df.shape[0]} แถว x {df.shape[1]} คอลัมน์")
print(f"คอลัมน์: {list(df.columns)}\n")
df.describe()
```

```python
# กรองข้อมูล (Filter)
print("--- จังหวัดที่ PM2.5 > 50 ---")
df[df["pm25"] > 50]
```

```python
# จัดเรียง
print("--- PM2.5 มากไปน้อย ---")
df.sort_values("pm25", ascending=False)[["province", "pm25"]].head()
```

```python
# groupby: สรุปตามกลุ่ม
print("--- PM2.5 เฉลี่ยตามภาค ---")
df.groupby("region")["pm25"].mean().sort_values(ascending=False)
```

## หัวข้อที่ 2 — matplotlib: การสร้างกราฟ
**เวลาโดยประมาณ: 12 นาที**

`matplotlib` เป็นไลบรารีสร้างกราฟและการแสดงผลข้อมูล (data visualization) ที่ใช้งานกว้างขวางที่สุดใน Python การเลือกประเภทกราฟที่เหมาะสมช่วยให้สื่อสารข้อมูลได้อย่างมีประสิทธิภาพ


```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# Bar Chart
colors = ['red' if pm > 50 else 'orange' if pm > 25 else 'green' for pm in df["pm25"]]

plt.figure(figsize=(12, 5))
plt.bar(df["province"], df["pm25"], color=colors)
plt.axhline(y=50, color='red', linestyle='--', label='Unsafe (>50)')
plt.axhline(y=25, color='green', linestyle='--', label='Safe (<=25)')
plt.ylabel('PM2.5')
plt.title('PM2.5 by Province')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: อุณหภูมิ vs PM2.5
plt.figure(figsize=(8, 5))
plt.scatter(df["temp"], df["pm25"], c=df["pm25"], cmap='RdYlGn_r', s=100, edgecolors='black')
plt.colorbar(label='PM2.5')
for i in range(len(df)):
    plt.annotate(df["province"][i], (df["temp"][i], df["pm25"][i]), fontsize=8)
plt.xlabel('Temperature (C)')
plt.ylabel('PM2.5')
plt.title('Temperature vs PM2.5')
plt.tight_layout()
plt.show()
```

## หัวข้อที่ 3 — การโหลดข้อมูลจากไฟล์ CSV
**เวลาโดยประมาณ: 6 นาที**

ในงานจริง ข้อมูลมักถูกเก็บในรูปแบบไฟล์ CSV (Comma-Separated Values) การใช้ `pd.read_csv()` และ `df.to_csv()` ช่วยให้อ่านและบันทึกข้อมูลได้สะดวก


```python
# สร้างข้อมูล PM2.5 รายเดือน (จำลอง)
import numpy as np
np.random.seed(42)

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
seasonal = [1.8, 1.9, 1.5, 1.0, 0.7, 0.5, 0.4, 0.5, 0.6, 0.8, 1.0, 1.5]

rows = []
for mi, m in enumerate(months):
    for prov, base in [("Bangkok",40),("ChiangMai",60),("Phuket",15),("KhonKaen",35),("Chonburi",25)]:
        pm = max(5, int(base * seasonal[mi] + np.random.randint(-10,10)))
        temp = round(28 + seasonal[mi]*3 + np.random.uniform(-2,2), 1)
        rows.append({"month":m, "province":prov, "pm25":pm, "temp_c":temp})

df_m = pd.DataFrame(rows)
df_m.to_csv("pm25_monthly.csv", index=False)
print(f"สร้างไฟล์ pm25_monthly.csv ({len(df_m)} แถว)")
df_m.head(10)
```

```python
# Line Chart: PM2.5 รายเดือนแยกจังหวัด
df2 = pd.read_csv("pm25_monthly.csv")

plt.figure(figsize=(12, 5))
for prov in df2["province"].unique():
    d = df2[df2["province"]==prov]
    plt.plot(d["month"], d["pm25"], marker='o', label=prov)

plt.axhline(y=50, color='red', linestyle='--', alpha=0.5, label='Unsafe')
plt.ylabel('PM2.5')
plt.title('Monthly PM2.5 by Province')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

## 📋 pandas + matplotlib Cheat Sheet

| `pd.read_csv("file.csv")` | โหลด CSV |

| `df.groupby("col").mean()` | สรุปกลุ่ม |
| `plt.bar(x, y)` | Bar Chart |
| `plt.plot(x, y)` | Line Chart |
| `plt.scatter(x, y)` | Scatter Plot |
| `plt.pie(sizes)` | Pie Chart |
