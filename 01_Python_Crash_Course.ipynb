## หัวข้อที่ 1 — `print()` ตัวแปร และชนิดข้อมูล
**เวลาโดยประมาณ: 5 นาที**

ตัวแปร (Variable) คือพื้นที่ในหน่วยความจำสำหรับเก็บข้อมูล ชนิดข้อมูลพื้นฐานใน Python มี 4 ประเภทหลัก ได้แก่ `str`, `int`, `float`, และ `bool`


```python
# สั่งให้แสดงผล
print("Hello Coding AI 2026!")
print("Python ใช้คำนวณได้:", 2 + 3, 10 / 3)
```

```python
# ตัวแปร = กล่องเก็บข้อมูล
name = "สมชาย"          # str  (ข้อความ)
age = 16                 # int  (จำนวนเต็ม)
height = 170.5           # float (ทศนิยม)
is_student = True        # bool (จริง/เท็จ)

# f-string: วิธีสร้างข้อความแบบง่าย
print(f"ชื่อ {name} อายุ {age} ปี สูง {height} cm")
print(f"ประเภท name: {type(name)}, age: {type(age)}")
```

```python
# รับค่าจากผู้ใช้ (ได้ string เสมอ → ต้องแปลง)
name = input("ชื่อ: ")
age = int(input("อายุ: "))
print(f"สวัสดี คุณ{name}! อีก {18 - age} ปีจะเข้ามหาลัย")
```

## หัวข้อที่ 2 — โครงสร้างเงื่อนไข `if-elif-else`
**เวลาโดยประมาณ: 5 นาที**

โครงสร้างเงื่อนไขทำให้โปรแกรมสามารถตัดสินใจได้อัตโนมัติตามค่าข้อมูลที่รับเข้ามา โดยใช้ตัวดำเนินการเปรียบเทียบ (`>`, `<`, `>=`, `<=`, `==`, `!=`) และตัวดำเนินการตรรกะ (`and`, `or`, `not`)


```python
# ตัวดำเนินการเปรียบเทียบ: > < >= <= == !=
# ตัวดำเนินการตรรกะ: and, or, not

pm25 = 65

if pm25 <= 25:
    print("🟢 อากาศดี")
elif pm25 <= 50:
    print("🟡 ปานกลาง")
elif pm25 <= 100:
    print("🟠 เริ่มมีผลต่อสุขภาพ")
else:
    print("🔴 อันตราย!")
```

```python
# รวมหลายเงื่อนไขด้วย and / or
temp = 36
pm25 = 80

if temp > 33 and pm25 > 50:
    print("🏠 ร้อน + อากาศแย่ → เปิดแอร์ ปิดหน้าต่าง")
elif temp <= 33 and pm25 <= 50:
    print("🍃 สบาย → เปิดหน้าต่างรับลม")
else:
    print("⚠️ ระวังอย่างใดอย่างหนึ่ง")
```

## หัวข้อที่ 3 — List และ `for` Loop
**เวลาโดยประมาณ: 10 นาที**

**List** คือโครงสร้างข้อมูลที่ใช้เก็บค่าหลายตัวในตัวแปรเดียว และ **`for` loop** คือคำสั่งที่ใช้วนซ้ำเพื่อประมวลผลข้อมูลทีละรายการ ซึ่งเป็นพื้นฐานสำคัญของการวิเคราะห์ข้อมูล


```python
# List เก็บข้อมูลหลายตัว
provinces = ["กรุงเทพ", "เชียงใหม่", "ภูเก็ต", "ขอนแก่น", "ชลบุรี"]
pm25_values = [55, 120, 25, 45, 30]

print("จำนวนจังหวัด:", len(provinces))
print("จังหวัดแรก:", provinces[0])     # นับจาก 0
print("PM2.5 สูงสุด:", max(pm25_values))
print("PM2.5 เฉลี่ย:", sum(pm25_values) / len(pm25_values))
```

```python
# for loop: วนทีละตัว
for province in provinces:
    print(f"📍 {province}")
```

```python
# ใช้ range() วนตาม index → เข้าถึงทั้ง 2 List พร้อมกัน
for i in range(len(provinces)):
    pm = pm25_values[i]
    status = "🟢" if pm <= 25 else "🟡" if pm <= 50 else "🟠" if pm <= 100 else "🔴"
    print(f"  {provinces[i]:<10} PM2.5 = {pm:>3} {status}")
```

```python
# สะสมค่าด้วย loop (Accumulator) — ใช้บ่อยมากในวิเคราะห์ข้อมูล
total = 0
danger_count = 0

for pm in pm25_values:
    total += pm
    if pm > 50:
        danger_count += 1

print(f"PM2.5 เฉลี่ย: {total / len(pm25_values):.1f}")
print(f"จังหวัดอากาศแย่: {danger_count} จังหวัด")
```

## หัวข้อที่ 4 — การสร้างและใช้งานฟังก์ชัน (Function)
**เวลาโดยประมาณ: 5 นาที**

**Function** คือชุดคำสั่งที่ตั้งชื่อไว้เพื่อให้เรียกใช้ซ้ำได้ การแบ่งโค้ดออกเป็นฟังก์ชันช่วยให้โค้ดอ่านง่าย บำรุงรักษาง่าย และลดการเขียนซ้ำซ้อน (DRY Principle)


```python
# สร้าง function = สร้างคำสั่งใหม่ที่ใช้ซ้ำได้
def check_air_quality(pm25):
    """รับค่า PM2.5 แล้วบอกสถานะ"""
    if pm25 <= 25:
        return "🟢 ดี"
    elif pm25 <= 50:
        return "🟡 ปานกลาง"
    elif pm25 <= 100:
        return "🟠 เริ่มแย่"
    else:
        return "🔴 อันตราย"

# ใช้งาน
print(check_air_quality(25))
print(check_air_quality(75))
print(check_air_quality(150))
```

```python
# function + loop + list = พลังเต็ม!
for i in range(len(provinces)):
    status = check_air_quality(pm25_values[i])
    print(f"{provinces[i]}: {status}")
```

## หัวข้อที่ 5 — Dictionary และโครงสร้างข้อมูลแบบตาราง
**เวลาโดยประมาณ: 5 นาที**

**Dictionary** คือโครงสร้างข้อมูลที่เก็บข้อมูลในรูปแบบคู่ key-value ซึ่งคล้ายกับตารางฐานข้อมูล และเป็นรากฐานของการทำงานกับ pandas DataFrame ในบทถัดไป


```python
# Dictionary = เก็บข้อมูลเป็นคู่ key: value
bangkok = {
    "province": "กรุงเทพ",
    "pm25": 55,
    "temp": 35.2,
    "humidity": 70
}

print(f"จังหวัด: {bangkok['province']}")
print(f"PM2.5: {bangkok['pm25']}")
print(f"อุณหภูมิ: {bangkok['temp']}°C")
```

```python
# List of Dictionaries — โครงสร้างเหมือนตาราง (เตรียมพร้อมสำหรับ pandas!)
weather_data = [
    {"province": "กรุงเทพ",    "pm25": 55,  "temp": 35.2},
    {"province": "เชียงใหม่",  "pm25": 120, "temp": 28.1},
    {"province": "ภูเก็ต",     "pm25": 25,  "temp": 31.5},
    {"province": "ขอนแก่น",   "pm25": 45,  "temp": 33.8},
    {"province": "ชลบุรี",     "pm25": 30,  "temp": 30.0},
]

print(f"{'จังหวัด':<12} {'PM2.5':>6} {'อุณหภูมิ':>8} {'สถานะ'}")
print("-" * 45)

for city in weather_data:
    status = check_air_quality(city["pm25"])
    print(f"{city['province']:<12} {city['pm25']:>5} {city['temp']:>7.1f}°C  {status}")
```

## สรุปสาระสำคัญ (Summary Cheat Sheet)

| หัวข้อ | ไวยากรณ์ (Syntax) | ตัวอย่าง |
|--------|-------------------|----------|
| ตัวแปรและชนิดข้อมูล | `name = value` | `x = 10`, `name = "AI"` |
| f-string | `f"ข้อความ {variable}"` | `f"ค่า {x} คือ {name}"` |
| รับค่าจากผู้ใช้ | `input("ข้อความ")` | `x = int(input("ใส่ค่า: "))` |
| เงื่อนไข | `if ... elif ... else` | `if x > 5:` ... `else:` |
| List | `[val1, val2, ...]` | `data = [1, 2, 3]`, `data[0]` |
| for loop | `for item in list:` | `for item in data:` |
| range loop | `for i in range(n):` | `for i in range(5):` |
| Function | `def name(params): return` | `def func(x): return x * 2` |
| Dictionary | `{"key": value}` | `d = {"name": "AI"}`, `d["name"]` |
